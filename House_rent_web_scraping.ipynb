```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

dt = []

for x in range(len(df)):
    link = df.iloc[x].values[1]
    url = link
    response = requests.get(url)
    print(response)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')

    # Initialize variables outside the loop
    Listing_Type = ''
    property_type = ''
    building_type = ''
    city = ''
    locality = ''
    area = ''
    rent = ''
    furnishing_status = ''
    num_rooms = ''
    num_bathrooms = ''
    covered_parking = ''
    view = ''
    floor_number = ''
    tower_block = ''
    additional_rooms = ''
    security_deposit = ''
    available_from = ''
    age_of_property = ''
    unit_no = ''
    total_floor_count = ''
    balcony = ''
    open_parking = ''
    deposit_amount = ''

    # Extract property information
    property_titles = soup.find_all(class_="projectLiData")
    
    for prop_title in property_titles:
        s = str(prop_title)
        s = s.split('<div class="projectLiData">')
        s = s[1].split('<strong>')

        label = s[0].strip()
        value = s[1].split('</strong>')[0].strip()

        # Update variables based on label
        if label == 'Listing Type':
            Listing_Type = value
        elif label == 'Property Type':
            property_type = value
        elif label == 'Building Type':
            building_type = value
        elif label == 'City':
            city = value
        elif label == 'Locality':
            locality = value
        elif label == 'Area':
            area = value
        elif label == 'Rent':
            rent = value
        elif label == 'Furnishing Status':
            furnishing_status = value
        elif label == 'Number of Rooms':
            num_rooms = value
        elif label == 'Number of Bathroom':
            num_bathrooms = value
        elif label == 'Covered Parking':
            covered_parking = value
        elif label == 'View':
            view = value
        elif label == 'Floor Number':
            floor_number = value
        elif label == 'Tower/Block':
            tower_block = value
        elif label == 'Additional Rooms':
            additional_rooms = value
        elif label == 'Security Deposit':
            security_deposit = value
        elif label == 'Available From':
            available_from = value
        elif label == 'Age of Property (Years)':
            age_of_property = value
        elif label == 'Unit No':
            unit_no = value
        elif label == 'Total Floor Count':
            total_floor_count = value
        elif label == 'Balcony':
            balcony = value
        elif label == 'Open/Uncovered Parking':
            open_parking = value
        elif label == 'Deposit amount':
            deposit_amount = value

    data = {
        'Listing Type': Listing_Type,
        'Property Type': property_type,
        'Building Type': building_type,
        'City': city,
        'Locality': locality,
        'Area': area,
        'Rent': rent,
        'Furnishing Status': furnishing_status,
        'Number of Rooms': num_rooms,
        'Number of Bathroom': num_bathrooms,
        'Covered Parking': covered_parking,
        'View': view,
        'Floor Number': floor_number,
        'Tower/Block': tower_block,
        'Additional Rooms': additional_rooms,
        'Security Deposit': security_deposit,
        'Available From': available_from,
        'Age of Property (Years)': age_of_property,
        'Unit No': unit_no,
        'Total Floor Count': total_floor_count,
        'Balcony': balcony,
        'Open/Uncovered Parking': open_parking,
        'Deposit amount': deposit_amount
    }
    dt.append(data)

# Create a DataFrame and save it to a CSV file
df = pd.DataFrame(dt)
df.to_csv('rent_dataset.csv')
```
